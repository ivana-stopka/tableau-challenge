## 2. Cleaning the Data

```python
# There is a old and new column structure. Will combine old df's together first and then 
# convert to new column structure prior to commencing cleaning.
frames_old = [jul_twenty_df,aug_twenty_df,sept_twenty_df,oct_twenty_df,nov_twenty_df,dec_twenty_df,jan_twentyOne_df]
old_df = pd.concat(frames_old,axis=0,join="outer",ignore_index=True)
old_df.head()
```

```python
# Drop, rename and re-order columns in old_df to match new structure (February 2021 onwards).

# Drop columns
old_df = old_df.drop(columns=["tripduration","birth year","gender"])
old_df.head()
```

```python
# Rename columns.
old_df = old_df.rename(columns={"bikeid":"ride_id","starttime":"started_at","stoptime":"ended_at","start station id":"start_station_id","start station name":"start_station_name","start station latitude":"start_lat","start station longitude":"start_lng","end station id":"end_station_id","end station name":"end_station_name","end station latitude":"end_lat","end station longitude":"end_lng","usertype":"member_casual"})
old_df.head()
```

```python
# Re-order columns
old_df = old_df[["ride_id","started_at","ended_at","start_station_name","start_station_id","end_station_name","end_station_id","start_lat","start_lng","end_lat","end_lng","member_casual"]]
old_df.head()
```

```python
# Check for NaN values
old_df.count()
```

```python
# Member_casual column values differ between old and new format. Change to match. Subscriber=member, Customer=casual
old_df["member_casual"] = old_df["member_casual"].replace({"Subscriber":"member","Customer":"casual"})
old_df["member_casual"].value_counts()
```

```python
old_df.head()
```

```python
# ...now onto the df's with the new column structure (feb 2021 onwards). Will combine new df's together first and then 
# match column structure to old_df prior to commencing cleaning.
frames_new = [feb_twentyOne_df,mar_twentyOne_df,apr_twentyOne_df,may_twentyOne_df,jun_twentyOne_df]
new_df = pd.concat(frames_new,axis=0,join="outer",ignore_index=True)
new_df.head()
```

```python
# Drop columns
new_df = new_df.drop(columns=["rideable_type"])
new_df.head()
```

```python
# Data types differ between the old and new df's for some columns due to amix of numerical and text station names and ids.
# Due to the columns not being required for calculations, i.e. they are just an identifier, can keep as object.
#old_df["start_station_id"].value_counts()
new_df["start_station_id"].value_counts()
```

```python
# Check for and remove any NaN values in the new_df
#new_df.count()
new_df=new_df.dropna(how="any")
new_df.count()
```

## 3. Combine the DataFrames and Export to CSV

```python
cleaned_frames = [old_df,new_df]
final_large_df = pd.concat(cleaned_frames,axis=0,join="outer",ignore_index=True)
```
